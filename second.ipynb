```python
def bank(n, k, arr):
    for i in range(k):
        max_value = max(arr)
        arr.insert(arr.index(max_value), max_value / 2)
        arr.insert(arr.index(max_value), max_value / 2)
        arr.pop(arr.index(max_value))
    return list(map(int, arr))
```

```python
n, k = list(map(int, input("Enter N and K").split()))
arr = []

for i in range(1,n+1):
    arr.append(int(input(f"Enter range {i} - {i+1}")))

bank(n, k, arr)
```
